```python
df2['n_moa']=df2.select_dtypes('int64').sum(axis = 1)
```

```python
df2['ativo_moa']=(df2['n_moa']!= 0)
```

```python
df2.head()
```

```python
df3 = pd.merge(df, df2[['id','n_moa', 'ativo_moa']], on= 'id')
```

```python
df3.head()
```

```python
df3.query('tratamento == "com_controle"')['ativo_moa'].value_counts()
```

```python
composto_principal = df3['composto'].value_counts().index[:5]
plt.figure(figsize=(10,8))
sns.boxplot(data=df3.query('composto in @composto_principal'), y='g-0', x='composto', hue='ativo_moa')
```

Usando Machine Learning para classica√ßao

```python
from sklearn.linear_model import LogisticRegression as lr
from sklearn.model_selection import train_test_split
```

```python
modelo_rlogistica = lr()
x=df3.select_dtypes('float64')
y=df3['ativo_moa']
x_treino,x_teste,y_treino,y_teste = train_test_split(x,y, test_size = 0.2)

modelo_rlogistica = lr(max_iter=1000)
modelo_rlogistica.fit(x_treino,y_treino)
modelo_rlogistica.score(x_teste,y_teste)
```

```python
from sklearn.dummy import DummyClassifier
from sklearn.metrics import accuracy_score
```

```python
modelo_dummy = DummyClassifier('most_frequent')
modelo_dummy.fit(x_treino, y_treino)
previsao_dummy = modelo_dummy.predict(x_teste)
accuracy_score(y_teste, previsao_dummy)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
```

```python
x=df3.select_dtypes('float64')
y=df3['ativo_moa']
x_treino,x_teste,y_treino,y_teste = train_test_split(x,y, test_size = 0.2)

modelo_arvore = DecisionTreeClassifier(max_depth=3)
modelo_arvore.fit(x_treino, y_treino)
modelo_arvore.score(x_teste, y_teste)
```

```python
fig, ax = plt.subplots(figsize = (15,10), facecolor='k')
tree.plot_tree(
    modelo_arvore,
    ax=ax,
    fontsize=10,
    rounded = True,
    filled = True,
    feature_names=x_treino.columns,
    class_names = ['N√£o Ativado', 'Ativado'])
plt.show
```

```python
x=df3.select_dtypes('float64')
y=df3['ativo_moa']
x_treino,x_teste,y_treino,y_teste = train_test_split(x,y,stratify = y, random_state = 376, test_size = 0.2)

teste = []
treino = []
for i in range(1,15):
  modelo_arvore = DecisionTreeClassifier(max_depth=i)
  modelo_arvore.fit(x_treino, y_treino)
  teste.append(modelo_arvore.score(x_teste, y_teste))
  treino.append(modelo_arvore.score(x_treino, y_treino))
```

```python
teste
```

```python
treino
```

```python
sns.lineplot(x = range(1,15), y = teste, label = 'teste')
sns.lineplot(x = range(1,15), y = treino, label = 'treino')
```
